```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_alcohol = pd.read_csv('data\who_alcohol.csv')
```

```python
for country in df_alcohol['Country Name'].unique():
    temp_data = df_alcohol[df_alcohol['Country Name']==country]
    plt.plot(temp_data['Year'], temp_data['Alcohol consumption'], label=country)
plt.title('Alcohol Consumption Over Time')
plt.xlabel('Year')
plt.ylabel('Liters of pure alcohol consumed')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_alcohol.loc[df_alcohol['Country Name']=='France', 'Country Name'] = 'FRA'
df_alcohol.loc[df_alcohol['Country Name']=='Estonia', 'Country Name'] = 'EST'
df_alcohol.loc[df_alcohol['Country Name']=='Czechia', 'Country Name'] = 'CZE'
df_alcohol.loc[df_alcohol['Country Name']=='South Africa', 'Country Name'] = 'SA'
df_alcohol.loc[df_alcohol['Country Name']=='California', 'Country Name'] = 'CA'
```

```python
df_alcohol[df_alcohol['Country Name'].isin(['FRA', 'EST', 'CZE', 'SA', 'CA'])].to_excel('data\combined_alcohol_excel.xlsx', index=False)
```
